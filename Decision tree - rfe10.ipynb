```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
data = pd.read_csv('rfe_dataset_10.csv')
```

```python
data
```

# Preprocesiranje

```python
X = data.drop('Class', axis = 1)
y = data[['Class']]
```

#  Podela podataka na trening i test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.2, random_state=42, stratify = y)
```

# Pravljenje modela i treniranje

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

# Tacnost modela

```python
accuracy_score(y_train, y_train_pred)
accuracy_score(y_test, y_test_pred)
```

```python
# Prikaz važnosti atributa u modelu
importance = model.feature_importances_
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
# Vizuelni prikaz stabla odlučivanja
_ = tree.plot_tree(model, filled = True)
```

```python
cm_train = confusion_matrix(y_train, y_train_pred)
```

```python
sns.heatmap(cm_train, annot=True, cmap='Reds')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
print(classification_report(y_train, y_train_pred))
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)
```

```python
sns.heatmap(cm_test, annot=True, cmap='Reds')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
y_train_prob = model.predict_proba(X_train)[:,1]
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
roc_auc_train = roc_auc_score(y_train, y_train_prob)

y_test_prob = model.predict_proba(X_test)[:,1]
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)
roc_auc_test = roc_auc_score(y_test, y_test_prob)

plt.figure(figsize=(8,6))
plt.plot(fpr_train, tpr_train, label='Train AUC = {:.2f}'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, label='Test AUC = {:.2f}'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], linestyle='--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
